```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
sda = pd.read_csv('fifa_eda_stats.csv')
sda
```

```python
sda.shape
```

```python
sda.replace("€", "", regex=True, inplace=True)  # Remove currency symbol
sda.fillna(np.nan, inplace=True)  # Replace empty values with NaN
```

```python
sda
```

```python
# Histogram of player ages
plt.figure(figsize=(10, 6))
sns.histplot(data=sda, x='Age', bins=20, kde=True)
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of player Overall vs Potential
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sda, x='Overall', y='Potential', alpha=0.5)
plt.title('Overall vs Potential')
plt.xlabel('Overall')
plt.ylabel('Potential')
plt.show()
```

```python
# Bar plot of top 10 Nationalities
plt.figure(figsize=(12, 6))
top_nationalities = sda['Nationality'].value_counts().head(10)
sns.barplot(x=top_nationalities.index, y=top_nationalities.values)
plt.title('Top 10 Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
